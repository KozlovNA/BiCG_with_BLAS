```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
# Replace 'your_file.csv' with the path to your CSV file
csv_file_path = './output/BCGSTAB_logs.csv'
data = pd.read_csv(csv_file_path)

# Check if the required columns exist
if 'k' in data.columns and 'res_2norm_rel' in data.columns:
    # Extract the columns
    x = data['k']
    y = data['res_2norm_rel']
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='b',alpha=0.5, label='BCGSTAB')
    
    plt.semilogy()
    
    # Show grid
    plt.grid()
    
else:
    print("The required columns 'k' and 'res_2norm' are not in the CSV file.")



# Load the CSV file
# Replace 'your_file.csv' with the path to your CSV file
csv_file_path = './output/BBCGSTAB_logs.csv'
data = pd.read_csv(csv_file_path)

# Check if the required columns exist
if 'k' in data.columns and 'res_max2norm_rel' in data.columns:
    # Extract the columns
    x = data['k']
    y = data['res_max2norm_rel']
    eps = 0.01*np.ones(len(x))
    # Create the plot
    # plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='r', alpha=0.5, label='BBCGSTAB')
    plt.plot(x, eps, label='threshold')
    # Adding titles and labels
    plt.title('Plot of res vs k BCGSTAB vs BBCGSTAB')
    plt.xlabel('k')
    plt.ylabel('res')
    plt.semilogy()
    
    # Show grid
    plt.grid()
    
    # Add legend
    plt.legend()

    plt.show()
    # Show the plot
else:
    print("The required columns 'k' and 'res_2norm' are not in the CSV file.")
```
