```python
import numpy as np
r0 = np.array([10, 3, 3])
pk = np.array([10, 3, 3])
sk = np.array([2.5295, -7.6964, -0.735252])
alpha = 118/695 
A = np.array([[2, 3, 5],
             [3, 7, 4],
             [1, 2, 2]])
vk = A@pk
print(A@sk)
tk = A@sk
omegak = np.dot(tk, sk)/np.dot(tk,tk)
print(omegak)
x = np.array([0.0,0.0,0.0])
x += pk*alpha + omegak*sk
print(x)
rkp1 = sk-omegak*tk
print(rkp1)
beta_rhs = -np.dot(r0,tk)
print(beta_rhs)
beta = beta_rhs/695
print(beta)
# pkp1 = rkp1-(omegak*vk) * beta + pk*beta
pkp1 = rkp1 + pk*beta
print (pk + pk*beta - omegak*vk*beta + rkp1)
```
