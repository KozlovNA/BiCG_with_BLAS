Проверим устойчивость алгоритма:
1) Будем решать с повышенной точностью
2) Каждый 10 шаг будем считать "честную невязку" как $r_k = b - Ax_k$
3) Будем выпонять промежуточные операции при подсчете скалярных произведений с повышенной точностью

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
# Replace 'your_file.csv' with the path to your CSV file
csv_file_path = './output/BCGSTABfloat_logs.csv'
data = pd.read_csv(csv_file_path)

# Check if the required columns exist
if 'k' in data.columns and 'res_2norm_rel' in data.columns:
    # Extract the columns
    x = data['k']
    y = data['res_2norm_rel']
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='b',alpha=0.5, label='BCGSTAB float')
    
    plt.semilogy()
    
    # Show grid
    plt.grid()
    
else:
    print("The required columns 'k' and 'res_2norm' are not in the CSV file.")



# Load the CSV file
# Replace 'your_file.csv' with the path to your CSV file
csv_file_path = './output/BCGSTABdouble_logs.csv'
data = pd.read_csv(csv_file_path)

# Check if the required columns exist
if 'k' in data.columns and 'res_2norm_rel' in data.columns:
    # Extract the columns
    x = data['k']
    y = data['res_2norm_rel']
    
    # Create the plot
    # plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='r', alpha=0.5, label='BCGSTAB double')
    
    # Adding titles and labels
    plt.title('Plot of res vs k')
    plt.xlabel('k')
    plt.ylabel('res')
    plt.semilogy()
    
    # Show grid
    plt.grid()
    
    # Add legend
    plt.legend()

    # Show the plot
    plt.show()
else:
    print("The required columns 'k' and 'res_2norm' are not in the CSV file.")
```
