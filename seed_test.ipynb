```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_file_path = './output/sbbcgsr/5 5.csv'

data = pd.read_csv(csv_file_path)

# Check if the required columns exist
if 'k' in data.columns and 'res_max2norm_rel' in data.columns:
    # Extract the columns
    x = np.array(data['k'])
    y = data['res_max2norm_rel']
    # Create the plot
    # plt.figure(figsize=(10, 6))
    eps = 0.01*np.ones(1700)
    plt.plot([0,850], [0.01, 0.01], label='threshold')
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='r', alpha=0.3)
    # Adding titles and labels
    plt.title('5 direction residuals and 5 other residuals')
    plt.xlabel('k')
    plt.ylabel(r'$max||res||_2 \; / \; max||res_0||_2 $')
    plt.semilogy()
    
    # Show grid
    plt.grid()
    
    # Add legend
    plt.legend()

    plt.show()
    # Show the plot
else:
    print("The required columns 'k' and 'res_2norm' are not in the CSV file.")
```
