```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_file_path = './output/metrics/omega_module_1rhs.csv'
data = pd.read_csv(csv_file_path)

if 'k' in data.columns and 'module' in data.columns:
    # Extract the columns
    x = np.array(data['k'])
    y = data['module']
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='m', alpha=0.5, label='1 правых частей')

#---------------------------------------------------------

csv_file_path = './output/metrics/omega_module_50rhs.csv'
data = pd.read_csv(csv_file_path)

if 'k' in data.columns and 'module' in data.columns:
    # Extract the columns
    x = np.array(data['k'])
    y = data['module']
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='g', alpha=0.5, label='50 правых частей')

#---------------------------------------------------------
csv_file_path = './output/metrics/omega_module_5rhs.csv'
data = pd.read_csv(csv_file_path)

if 'k' in data.columns and 'module' in data.columns:
    # Extract the columns
    x = np.array(data['k'])
    y = data['module']
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='b', alpha=0.5, label='5 правых частей')

#---------------------------------------------------------

csv_file_path = './output/metrics/omega_module_20rhs.csv'
data = pd.read_csv(csv_file_path)

if 'k' in data.columns and 'module' in data.columns:
    # Extract the columns
    x = np.array(data['k'])
    y = data['module']
    plt.plot(x, y, marker='o', markersize=3, linestyle='-', color='r', alpha=0.5, label='20 правых частей')

    # Adding titles and labels
    plt.title(r' $|\omega_k|$ модифицированный Bl-BiCGStab')
    plt.xlabel('k')
    plt.ylabel(r'$max||res||_2 \; / \; max||res_0||_2 $')
    plt.semilogy()
    
    # Show grid
    plt.grid()
    
    # Add legend
    plt.legend()
    plt.savefig('./output/bbcgsr/plots/acceleration_15_rhs.pdf')
    plt.show()
    # Show the plot
else:
    print("The required columns 'k' and 'res_2norm' are not in the CSV file.")
```
